```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
edades = [ 21 ,20 ,19 ,42, 22 ,23 ,38 ,19 ,43 ,19 ,33 ,22 ,26 ,25 ,22 ,25 ,28, 21 ,21, 27, 38,
          20, 20, 30, 19, 22 ,34 ,35 ,34 ,24 ,40 ,19 ,42 ,26 ,23 ,20 ,20 ,33, 29, 31]
nro_intervalos = 8
_min, _max = min(edades), max(edades)
amplitud = (_max - _min) / nro_intervalos
extremos = [_min + i*amplitud for i in range(nro_intervalos+1)]
frecuencia, bins = np.histogram(edades, extremos)
print(extremos, len(extremos))
print(frecuencia, len(frecuencia), bins, len(bins))
```

```python
D = {
    'Marcas de Clase': [(bins[i]+bins[i+1]) / 2 for i in range(nro_intervalos)],
    'Frecuencia. (fi)': frecuencia
}
df = pd.DataFrame(D)
N = df['Frecuencia. (fi)'].sum(); df['Frecuencia Relativa. (fr)'] = df['Frecuencia. (fi)']/N
df['Frecuencia Relativa Acumulada. (fr ac)'] = df['Frecuencia Relativa. (fr)'].cumsum()
df.interpolate()
```

```python
fig1, ax1 = plt.subplots()
ax1.bar(df['Marcas de Clase'], df['Frecuencia Relativa. (fr)'], width=amplitud*0.95)
ax1.plot(df['Marcas de Clase'], df['Frecuencia Relativa. (fr)'], color='red', marker='o', label='Poligono Frecuencia')
ax1.set_xlabel('Intervalos'); ax1.set_ylabel('Frecuencia Relativa. (fi)'); ax1.set_title('Histograma de Frecuencias Relativas')
ax1.legend(); ax1.grid()
```

```python
fig2, ax2 = plt.subplots()
ax2.bar(df['Marcas de Clase'], df['Frecuencia Relativa Acumulada. (fr ac)'], width=amplitud*0.95)
ojiva_x, ojiva_y = extremos, [0]
ojiva_y.extend(df['Frecuencia Relativa Acumulada. (fr ac)'])
print(ojiva_x, len(ojiva_x), ojiva_y, len(ojiva_y), sep='\n')

ax2.plot(ojiva_x, ojiva_y, color='red', marker='o', label='Ojiva')
ax2.plot([22]*40, np.linspace(0,1,40),color='black', label='22 años')
ax2.plot([35]*40, np.linspace(0,1,40),color='black', label='35 años')

ax2.legend(); ax2.grid()
```

```python
edades.sort()
print(edades)
```

```python
c = 0
for e in edades:
    if e < 22:
        c += 1
c_rel = c / N
print(f'Hay {c} mujeres con menos de 22 años\nEl {c_rel} tienen menos de 22 años')

c = 0
for e in edades:
    if e > 35:
        c += 1
c_rel = c / N
print(f'Hay {c} mujeres con mas de 35 años\nEl {c_rel} tienen mas de 35 años')

```

```python
c = int(N/4)
print('El 25% de mujeres son:', c)

#recordando indices empizan de 0
#ordenar para evitar while y logica de mayor o menor
edades.sort()
edad_maxima = edades[c-1]
print(f'El 25% de menor edad tienen a lo sumo {edad_maxima} años')

edades.sort(reverse=True)
edad_minima = edades[c-1]
print(f'El 25% de mayor edad tienen como poco {edad_minima} años')
```
